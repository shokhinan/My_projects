### ***'balcony'***

Заполним нулями пропущенные значения ***в колонке "balcony"*** , а также изменим тип на int(не bool, тк может быть > 1 балкона)

```python
data['balcony'] = data['balcony'].fillna(0)
data['balcony'].value_counts()
```

### ***'ceiling_height'***

В данной колонке прослеживается аномалия, очень большая высота потолков (до 100м). Есть предположение, что пользователи не там поставили запятую, поэтому появилась высота 27 метров и более. Поэтому заменим высоту потолков более 22 метров на эту высоту, уменьшенную в 10 раз.

```python
data.loc[data['ceiling_height'] >= 22, 'ceiling_height'] /= 10
data.sort_values('ceiling_height').head(30)
```

Коффициент корреляции 'ceilling_height' очень  маленький, поэтому может заполнить данную колонку медианными значениями

```python
data['ceiling_height'] = data['ceiling_height'].fillna(data['ceiling_height'].median())
data.isna().sum()
data['ceiling_height'].describe()
```

По этим данным смущают значения менее 2 метров и более 4, их совсем немного, потому удалим

```python
data = data.loc[(data['ceiling_height'] >= 2)&(data['ceiling_height'] <= 4)]
data['ceiling_height'].describe()
```

Для самых часто встречающихся мест попробуем найти дубликаты

```python
def in_group(row):
    try:
        if 'анкт' in row:
            return 'Санкт_Петербург'
        elif "Зелёный Бор" in row:
            return 'поселок Зеленый Бор'
        elif "урино" in row:
            return "Мурино"
        elif "ябово" in row:
            return "Рябово"
        elif "ушар" in row:
            return "Шушары"
        elif "ельман" in row:
            return "Тельман"
        elif "Всеволжск" in row:
            return "Всеволжск"
        else:
            return row
    except:
        return row
    
data['local_big'] = data['locality_name'].apply(in_group)
data
```

***Тип этажа квартиры.*** создаем функцию how_floor, которая будет определять этаж (first, last, other)

```python
def floor(row):
    floors_total = row['floors_total']
    floor = row['floor']
    try:
        if floor == 1:
            return 'first'
        elif floor == floors_total:
            return 'last'
        elif 1 < floor < floors_total:
            return 'other'
    except:
        return 'Проверьте правильность этажа'
    
data['floor_category'] = data.apply(floor, axis = 1)
data
```

### Как стоимость объектов зависит от расстояния до центра города.

Протяжённость Санкт-Петербурга по линии север-юг оценивается примерно в 44 км. Поэтому уберем все значения > 22 км. Это уже будет не Санкт-Петербург

```python

data_spb = data[(data['locality_name'] == 'санкт-петербург')&(data['cityCenters_nearest_km'] <= 22)].dropna(subset=['cityCenters_nearest_km']) # удалим те строки, где неизвестно расстояние
data_spb = data_spb[['last_price', 'cityCenters_nearest_km']]#составили таблицу с ценой объекта и расстоянием до центра
data_spb['cityCenters_nearest_km'] = data_spb.loc[:, 'cityCenters_nearest_km'] #добавляем столбец с км
data_spb['cityCenters_nearest_km'] = data_spb['cityCenters_nearest_km'].astype(int) #меняем тип на целый, чтобы потом группировать
data_spb['mean_price_for_km'] = data['price_square_meter'].mean() / data_spb['cityCenters_nearest_km'] #средняя цена для каждого км
data_spb.groupby('cityCenters_nearest_km')['mean_price_for_km']

data_spb = data_spb.drop_duplicates().sort_values(by='cityCenters_nearest_km')

data_spb
```

```python
data_new_spb = data_spb.pivot_table(index = 'cityCenters_nearest_km', values = 'last_price', aggfunc = ['median'])
data_new_spb.columns = ['median']
data_new_spb.sort_values('median', ascending = False)
data_new_spb.plot(y='median', kind='bar')
None
data_new_spb
```
